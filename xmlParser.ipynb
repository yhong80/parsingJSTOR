```python
path1 = './title-group'
path2 = './abstract/p'
root=tree.getroot()
content={}
for bookpart in root.iter('book-part-meta'):
    root = bookpart
    
    title = root.find(path1)
    name = ""
    for t in title:
        name = name + t.text + " : "
    #print(name)
    
    abstract = root.find(path2)
    if (abstract is None):
        abst = ""
    else:
        abst = abstract.text
    content[name]=abst

print(content.keys())
    
            
```

```python
root=tree.getroot()
for b in root.iter('abstract'):
    root = b
    for i in b.iter('p'):
        print(i.text)
    print("\n")
```
